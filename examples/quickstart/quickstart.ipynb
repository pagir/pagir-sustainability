## 2. Run SustainAction â†’ Get Recommendations

````python
engine = SustainActionEngine('starter_actions.yaml')
actions = engine.match_actions('plant_energy_baseline.csv')

total_kwh = sum(a['estimated_savings_kwh'] for a in actions)
total_co2e = sum(a['estimated_savings_co2e'] for a in actions)

print('=== AJITH\'S FACTORY SAVINGS PLAN ===')
print(f'ðŸŽ¯ TOTAL: {total_kwh:,} kWh | {total_co2e} tCOâ‚‚e/year')
print(f'ðŸ“Š = {total_kwh/1500000*100:.1f}% of baseline energy use')
print()
````

## 3. Results Table

````python
import pandas as pd

df = pd.DataFrame(actions)
df = df[['name', 'asset_count', 'assets', 'estimated_savings_kwh', 'estimated_savings_co2e']]
df['estimated_savings_kwh'] = df['estimated_savings_kwh'].round(0).astype(int)
df['estimated_savings_co2e'] = df['estimated_savings_co2e'].round(1)
df
````
